```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
gcd = pd.read_csv(r'C:\Users\German\Desktop\german_credit_data.csv')
```

```python
gcd.head(10)
```

```python
#jointplot
sns.jointplot(x='Age', y='Credit amount', data=gcd)
```

```python
#displot
sns.distplot(gcd['Duration'], kde=False, color='green')
```

```python
#boxplot
sns.boxplot(x='Housing', y="Age", data=gcd)
```

```python
#swarmplot
sns.swarmplot(x='Housing', y="Age", data=gcd, palette='plasma')
```

```python
#countplot
sns.countplot(x="Sex", data=gcd)
```

```python
#heatmap
sns.heatmap(gcd.corr(), cmap='winter')
plt.title('gcd.corr()')
```

```python
#Facetgrid
grid = sns.FacetGrid(data=gcd, col='Housing')
grid.map(plt.hist, 'Credit amount')
```
